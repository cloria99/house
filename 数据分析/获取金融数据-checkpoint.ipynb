## 1.1马应龙公司

```python
import pandas_datareader.data as web

# 读取德国DAX指数股价信息
DAX = web.DataReader(name="600993.SS",data_source='yahoo',start='2019-6-1')
DAX.info()
```

```python
DAX.tail()
```

```python
import matplotlib.pyplot as plt
# DAX.loc[:,['Close']]
DAX['Close'].plot(figsize=(8,5)) 
#plt.xlim("2019-6-1", "2020-6-24")
```

```python
%%time
DAX['Ret_Loop'] = 0.0
for i in range(1,len(DAX)):
    DAX['Ret_Loop'][i] = np.log(DAX['Close'][i] / DAX['Close'][i-1])#根据每日收盘价返回对数收益率
```

```python
DAX[['Close','Ret_Loop']].tail()
```

```python
%time DAX['Return'] = np.log(DAX['Close'] / DAX['Close'].shift(1))

DAX[['Close','Ret_Loop','Return']].tail()
```

```python
del DAX['Ret_Loop']
DAX
```

```python
DAX['42d'] = DAX['Close'].rolling(window=42,center=False).mean()
DAX['252d'] = DAX['Close'].rolling(window=252,center=False).mean()

DAX[['Close','42d','252d']].tail()
```

```python

DAX[['Close','Return']].plot(subplots=True,style='b',figsize=(12,5))
plt.xlim("2019-6-1", "2020-6-24")
#DAX.plot.scatter("2020-1-1","2020-06-24")
```

```python
DAX[['Close','42d']].plot(figsize=(8,5))
plt.xlim("2019-6-1", "2020-6-24")
```

从文件中获取的信息

```python
import baostock as bs 
import pandas as pd
#### 登陆系统 #### 
lg = bs.login() # 显示登陆返回信息 
print('login respond error_code:'+lg.error_code) 
print('login respond error_msg:'+lg.error_msg) 
#### 获取沪深A股历史K线数据 #### 

# 详细指标参数，参见“历史行情指标参数”章节
rs = bs.query_history_k_data("sh.600993", "date,code,open,high,low,close,preclose,volume,amount,adjustflag,turn,tradestatus,pctChg,isST",
                             start_date='2019-06-01', end_date='2020-6-24', frequency="d", adjustflag="3")
print('query_history_k_data respond error_code:'+rs.error_code)
print('query_history_k_data respond error_msg:'+rs.error_msg)

#### 打印结果集 #### 
data_list = [] 
while (rs.error_code == '0') & rs.next():
    # 获取一条记录，将记录合并在一起
    data_list.append(rs.get_row_data()) 
result = pd.DataFrame(data_list, columns=rs.fields) 
    #### 结果集输出到csv文件 #### 
result.to_csv(r"C:\Users\Lenovo\Desktop\1.csv", index=False) 
print(result) 
#### 登出系统 #### 
bs.logout()
```

```python
import datetime
import csv
import matplotlib.pyplot as plt
import mpl_finance as mpf
import matplotlib.dates as mpd
import numpy as np

str2date =lambda x: mpd.date2num(datetime.datetime.strptime(x, '%Y/%m/%d').date())
data_list = np.loadtxt(r"C:/Users/Lenovo/Desktop/1.csv",str,delimiter=",", skiprows=1)
columns = data_list[0]
quotes = [[str2date(d[0])] + [float(v) for v in d[1:-1]] for d in data_list[1:]]
fig,ax = plt.subplots(figsize=(12,5), facecolor=(0.5, 0.5, 0.5))
fig.subplots_adjust(bottom=0.2)
mpf.candlestick_ohlc(ax,quotes,width=0.4,colorup='r',colordown='g')
plt.grid(False)
ax.xaxis_date()
ax.autoscale_view()
plt.setp(plt.gca().get_xticklabels(), rotation=30)
plt.title("mayinglong Index")
plt.xlabel("Date")
plt.ylabel("close")
plt.show()
```
